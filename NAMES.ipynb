```python
df['Python'].replace(np.NaN,value=df['Python'].mean(),inplace=True)
df['Machine Learning'].replace(np.NaN,value=df['Machine Learning'].mean(),inplace=True)
df['Age'].replace(np.NaN,value=df['Age'].mean(),inplace=True)
df['AI'].replace(np.NaN,value=df['AI'].mean(),inplace=True)
df['NLP'].replace(np.NaN,value=df['NLP'].mean(),inplace=True)
df['Cv'].replace(np.NaN,value=df['Cv'].mean(),inplace=True)
```
