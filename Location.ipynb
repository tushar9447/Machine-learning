```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Location.csv")
```

```python
df
```

```python
def myfun(x):
    return x.split(",")[0]

def myfun1(x):
    return x.split(",")[1]
df['city']=df['Location'].apply(myfun)
df['state']=df['Location'].apply(myfun1)
```

```python
df
```

```python
def myfun1(x):
    if x>=18:
        return "adult"
    else:
        return "teen"
    
df['type']=df['Age'].apply(myfun1)
```

```python
df
```

```python
def myfun3(x):
    return int(x.replace("$"," "))

df['Salary_dollar']=df['Salary'].apply(myfun3)
```

```python
df
```

```python
def myfun4(x):
    return int(x.replace("$",""))*81

df['Salary_rs']=df['Salary'].apply(myfun4)
```
