# Encoding :


```python
# Distinguishing the Numerical & Categorial values 
df_numerical = df.select_dtypes(['int64','float64'])
df_category = df.select_dtypes(object)
```

# ONE HOT ENCODING

#### 1) Find unique catgory
#### 2) Apply get_dummies() on the column
#### 3) Reinitialize to the DataFrame

```python
pd.get_dummies(df_category['fuel-type'])
```

# LabelEncoding

#### 1) import LabelEncoder
#### 2) create Instance/Object of LabelEncoder class
#### 3) apply obj.fit_transform() on particular feature/column
#### 4) Reinitialize to the DataFrame

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
LE.fit_transform(df_category['fuel-type'])
```

```python
for feature_x in df_category:
    LE = LabelEncoder()
    df_category[feature_x] = LE.fit_transform(df_category[feature_x])
```

```python
df_category.head()
```

```python
new_df = pd.concat([df_numerical,df_category],axis=1)
```

```python
new_df.head()
```

```python
new_df.info()
```
